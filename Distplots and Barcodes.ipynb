# Distplot

```python
sns.distplot(tips_df['size'])
plt.show()
```

```python
sns.distplot(tips_df['total_bill'])
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],bins=10)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],hist=False)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],kde=False)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],rug=True)
plt.show()
```

```python
from scipy.stats import norm 
sns.distplot(tips_df['total_bill'],fit=norm,kde=False)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],color='r',fit=norm,kde=False)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],color='r',fit=norm,kde=False,vertical=True)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],color='r',fit=norm,kde=False,axlabel='Bill Paid')
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],color='r',fit=norm,kde=False,label='Total bill')
plt.legend()
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],hist_kws={'color':'y','linewidth':2 ,'edgecolor':'r','alpha':0.7})
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],rug=True,hist_kws={'color':'y','linewidth':2 ,'edgecolor':'r','alpha':0.7},kde_kws={'color':'b','linewidth':4 ,'alpha':0.7},rug_kws={'color':'y','linewidth':2 ,'edgecolor':'g','alpha':0.7})
plt.show()
```
