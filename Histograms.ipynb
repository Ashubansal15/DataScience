```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
miet_student_age=np.random.randint(18,45,100)
mit_student_age=np.random.randint(15,40,100)
```

```python
#hist
```

```python
plt.hist(miet_student_age)
plt.show()
```

```python
plt.hist(miet_student_age)
plt.title("Miet Histogram")
plt.xlabel("Age")
plt.ylabel("Number of students")
plt.show()
```

```python
plt.hist(miet_student_age,bins=[15,20,25,30,35,40,45])
plt.title("Miet Histogram")
plt.xlabel("Age")
plt.ylabel("Number of students")
plt.show()
```

```python
plt.hist(miet_student_age,bins=[15,20,25,30,35,40,45],rwidth=0.8)
plt.title("Miet Histogram")
plt.xlabel("Age")
plt.ylabel("Number of students")
plt.show()
```

```python
plt.hist(miet_student_age,bins=[15,20,25,30,35,40,45],rwidth=0.8,histtype='barstacked')
plt.title("Miet Histogram")
plt.xlabel("Age")
plt.ylabel("Number of students")
plt.show()
```

```python
plt.hist(miet_student_age,bins=[15,20,25,30,35,40,45],rwidth=0.8,histtype='barstacked',orientation='horizontal')
plt.title("Miet Histogram")
plt.xlabel("Age")
plt.ylabel("Number of students")
plt.show()
```

```python
plt.hist(miet_student_age,bins=[15,20,25,30,35,40,45],rwidth=0.8,color='g')
plt.title("Miet Histogram")
plt.xlabel("Age")
plt.ylabel("Number of students")
plt.show()
```

```python
plt.hist(miet_student_age,bins=[15,20,25,30,35,40,45],rwidth=0.8,color='g')
plt.title("Miet Histogram")
plt.xlabel("Age")
plt.ylabel("Number of students")
plt.legend("age",loc=2)
plt.show()
```

```python
plt.hist(miet_student_age,bins=[15,20,25,30,35,40,45],rwidth=0.8,color='g',label='miet age')
plt.title("Miet Histogram")
plt.xlabel("Age")
plt.ylabel("Number of students")
plt.legend(loc=2)
plt.show()
```

```python
plt.hist(miet_student_age,bins=[15,20,25,30,35,40,45],rwidth=0.8,color='g',label='miet age')
plt.hist(mit_student_age,bins=[15,20,25,30,35,40,45],rwidth=0.8,color='r',label='mit age')
plt.title("Miet Histogram")
plt.xlabel("Age")
plt.ylabel("Number of students")
plt.legend(loc=2)
plt.show()
```

```python
plt.hist([miet_student_age,mit_student_age],bins=[15,20,25,30,35,40,45],rwidth=0.8,color=['g','r'],label=['miet age','mit_age'])
#plt.hist(mit_student_age,bins=[15,20,25,30,35,40,45],rwidth=0.8,color='r',label='mit age')
plt.title("Miet Histogram")
plt.xlabel("Age")
plt.ylabel("Number of students")
plt.legend(loc=2)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.hist([miet_student_age,mit_student_age],bins=[15,20,25,30,35,40,45],rwidth=0.8,color=['g','r'],label=['miet age','mit_age'])
#plt.hist(mit_student_age,bins=[15,20,25,30,35,40,45],rwidth=0.8,color='r',label='mit age')
plt.title("Miet Histogram")
plt.xlabel("Age")
plt.ylabel("Number of students")
plt.legend(loc=2)

plt.show()
```
