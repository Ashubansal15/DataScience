```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("googleplaystore.csv",nrows=100)
df
```

```python
df.describe()
```

```python
x=df['Rating']
y=df['Reviews']
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.xlabel("Rating")
plt.ylabel("No of Reviews")
plt.title("Review vs Rating")
plt.scatter(x,y,color='r')
```

```python
plt.figure(figsize=(8,8))
plt.xlabel("Rating")
plt.ylabel("No of Reviews")
plt.title("Review vs Rating")
plt.scatter(x,y,color='r',marker='^',s=40,alpha=0.5)
```

```python
plt.figure(figsize=(8,8))
plt.xlabel("Rating")
plt.ylabel("No of Reviews")
plt.title("Review vs Rating")
plt.scatter(x,y,color='r',marker='^',s=40,alpha=0.5,linewidths=5)
```

```python
plt.figure(figsize=(8,8))
plt.xlabel("Rating")
plt.ylabel("No of Reviews")
plt.title("Review vs Rating")
plt.scatter(x,y,color='r',marker='^',s=40,alpha=0.5,linewidths=5)
```

```python
classes=['py','c','c++','ai','p']
batch1=[45,15,30,85,38]
```

```python
plt.pie(batch1,labels=classes)
plt.show()
```

```python
explode=[0,0,0.2,0,0]
plt.pie(batch1,labels=classes,explode=explode)
plt.show()
```

```python
color=['g','m','y','c','k']
plt.pie(batch1,labels=classes,colors=color)
plt.show()
```

```python
color=['g','m','y','c','b']
plt.pie(batch1,labels=classes,colors=color,autopct='%0.1F%%')
plt.show()
```

```python
color=['g','m','y','c','b']
plt.pie(batch1,labels=classes,colors=color,autopct='%0.1F%%',shadow=True)
plt.show()
```

```python
plt.pie(batch1,labels=classes,colors=color,autopct='%0.1F%%',shadow=True,radius=1.4)
plt.show()
```

```python
plt.pie(batch1,labels=classes,colors=color,autopct='%0.1F%%',shadow=True,radius=1.4,startangle=250)
plt.show()
```

```python
textprops={'fontsize':15}
color=['g','m','y','c','b']
plt.pie(batch1,labels=classes,colors=color,autopct='%0.1F%%',shadow=True,textprops=textprops,radius=1.5)
plt.show()
```
